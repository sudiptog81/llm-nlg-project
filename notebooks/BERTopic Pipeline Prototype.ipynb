```python
df_climate = pd.read_csv(f'{DATASET_PATH}/Indian climate/data_clean_doc.csv')
doc_climate = df_climate['Sentences'].tolist()
```

```python
df_economy = pd.read_csv(f'{DATASET_PATH}/Indian economy/data_clean_doc.csv')
doc_economy = df_economy['Sentences'].tolist()
```

```python
df_infrastructure = pd.read_csv(f'{DATASET_PATH}/Indian infrastructure/data_clean_doc.csv')
doc_infrastructure = df_infrastructure['Sentences'].tolist()
```

```python
df_defense = pd.read_csv(f'{DATASET_PATH}/Indian defense/data_clean_doc.csv')
doc_defense = df_defense['Sentences'].tolist()
```

```python
doc = doc_economy + doc_climate + doc_infrastructure + doc_defense
doc = [" ".join([w.lower() for w in nltk.word_tokenize(d) if w not in stop_words and w.isalpha()]) for d in doc_climate]
```

```python
freqs = []

for d in [doc_economy, doc_climate, doc_infrastructure, doc_defense]:
  d = [" ".join([w.lower() for w in nltk.word_tokenize(_d)
      if w not in stop_words and w.isalpha()]) for _d in d]
  topic_model = BERTopic(verbose=True, calculate_probabilities=True,
                        embedding_model="all-MiniLM-L6-v2")
  topics, probs = topic_model.fit_transform(d)
  freq = topic_model.get_topic_info()
  freqs.append(freq)
```
