```python
import json
from scholarly import scholarly

organic = scholarly.search_pubs("Indian infrastructure")
titles = []
cnt = 1

with open('/content/drive/MyDrive/DUCSResearch/Bias in LLMs/colab_notebooks/bias_llm/Indian infra prompts.txt', 'w') as f:
  for o in organic:
    if cnt == 1000:
      break
    print(o['bib']['title'])
    f.write(o['bib']['title'] + '\n')
    cnt += 1
```

```python
TOPICS = ['economy', 'climate', 'defence', 'infra']
for topic in TOPICS:
  s = set()
  with open(f'/content/drive/MyDrive/DUCSResearch/Bias in LLMs/colab_notebooks/bias_llm/Indian {topic} prompts.txt', 'r') as f:
    for line in f.readlines():
      l = line.strip().lower()
      if topic[:-3] in l and 'india' in l:
        s.add(l)

  with open(f'/content/drive/MyDrive/DUCSResearch/Bias in LLMs/colab_notebooks/bias_llm/Indian {topic} prompts-cleaned.txt', 'w') as f:
    for line in s:
      f.write(line + '\n')
```
